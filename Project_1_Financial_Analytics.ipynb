```python
data.columns = data.columns.str.strip().str.replace(' ', '_').str.replace('–', '-').str.lower()
print(data.columns)
```

```python
print("Market Capitalization Statistics:")
print(data['mar_cap_-_crore'].describe())

print("\nQuarterly Sales Statistics:")
print(data['sales_qtr_-_crore'].describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['mar_cap_-_crore'], bins=30, kde=True)
plt.title('Distribution of Market Capitalization',fontsize = 16)
plt.xlabel('Market Capitalization in Crores',fontsize = 14)
plt.ylabel('Frequency',fontsize = 14)
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['sales_qtr_-_crore'], bins=30, kde=True)
plt.title('Distribution of Quarterly Sales', fontsize=16)
plt.xlabel('Quarterly Sales in Crores', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sales_qtr_-_crore', y='mar_cap_-_crore', data=data)
sns.regplot(x='sales_qtr_-_crore', y='mar_cap_-_crore', data=data, scatter=False, color='red')
plt.title('Market Capitalization vs Quarterly Sales', fontsize=16)
plt.xlabel('Quarterly Sales in Crores', fontsize=14)
plt.ylabel('Market Capitalization in Crores', fontsize=14)
plt.grid(True)
plt.show()
```

```python
correlation = data['mar_cap_-_crore'].corr(data['sales_qtr_-_crore'])
print(f"Correlation between Market Capitalization and Quarterly Sales: {correlation}")
```

```python
# Segmentation Analysis
data['market_cap_segment'] = pd.qcut(data['mar_cap_-_crore'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
# Mean Quarterly Sales for each Market Cap Segment
segment_analysis = data.groupby('market_cap_segment')['sales_qtr_-_crore'].mean()
print(segment_analysis)
```

```python
# Bar plot for segment analysis
plt.figure(figsize=(10, 6))
segment_analysis.plot(kind='bar', color='skyblue')
plt.title('Average Quarterly Sales by Market Cap Segment', fontsize=16)
plt.xlabel('Market Cap Segment', fontsize=14)
plt.ylabel('Average Quarterly Sales in Crores', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Regression Analysis
from sklearn.linear_model import LinearRegression

# Prepare the data
X = data['sales_qtr_-_crore'].values.reshape(-1, 1)
y = data['mar_cap_-_crore'].values
```

```python
# Create and fit the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Coefficients of the regression model
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")

# Predicting Market Cap
data['predicted_mar_cap'] = model.predict(X)
```

```python
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sales_qtr_-_crore', y='mar_cap_-_crore', data=data, label='Actual')
sns.lineplot(x='sales_qtr_-_crore', y='predicted_mar_cap', data=data, color='red', label='Regression Line')
plt.title('Market Capitalization vs Quarterly Sales with Regression Line', fontsize=16)
plt.xlabel('Quarterly Sales in Crores', fontsize=14)
plt.ylabel('Market Capitalization in Crores', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
